```python
import matplotlib.pyplot  as plt
x1=range(0,10)
y1=[10,13,5,40,30,60,70,12,55,25]

x2=range(0,10)
y2=[5,8,0,30,20,40,50,10,40,15]

plt.plot(x1,y1,'ro-',label='Frist line',linewidth=3,markerfacecolor='blue',markersize=8)
plt.plot(x2,y2,'g*-',label='Second line',linewidth=3,markerfacecolor='yellow',markersize=8)
```

```python
#matplotlib.rcParams['font.sans-serif']=['SimHei']
#matplotlib.rcParams['axes.unicode_minus']=False

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False

label_list=['2014','2015','2016','2017']  ##横坐标刻度显示值
num_list1=[20,30,15,35]
num_list2=[15,30,40,20]
x1=list(range(len(num_list1)))
x2=[i+0.4 for i in x1]

plt.bar(x1,height=num_list1,width=0.4,align='edge',color='red',label="一部门")
plt.bar(x2,height=num_list2,width=0.4,align='edge',color='green',label="二部门")

plt.ylim(0,50)
plt.ylabel("数量")

plt.xticks([index+0.4 for index in x1],label_list)
plt.xlabel("年份")

plt.title("公司")
plt.legend()

plt.show()
```

```python
price=[39.5,39.9,45.4,38.9,33.34]

plt.barh(range(5),price,height=0.7,color='steelblue',alpha=0.8)

plt.yticks(range(5),['亚马逊','当当网','中国图书网','京东','天猫'])
plt.xlim(30,47)
plt.xlabel("价格")
plt.ylabel("商家")
plt.title("不同平台图书价格")

for x,y in enumerate(price):
    plt.text(y+0.2,x-0.1,'%s'%y)

plt.show()
```

```python
import numpy as np
courses=['C++','Python','高数','大学英语','软件工程','组成原理','数字图像处理','计算机图形学']
scores=[80,95,78,85,45,65,80,60]

dataLength=len(scores)

angles=np.linspace(0,
                   2*np.pi,
                   dataLength,
                   endpoint=False)

scores.append(scores[0])
angles=np.append(angles,angles[0])

plt.polar(angles,
          scores,
          'rv--',
          linewidth=2)

plt.thetagrids(angles[:8]*180/np.pi,
               courses
              )

plt.show()
```

```python
X=np.arange(-5,5.25,0.25)
Y=np.arange(-5,5.25,0.25)
X,Y=np.meshgrid(X,Y)
print(X,Y)

Z=np.sqrt(X**2+Y**2)
print(Z)

ax=plt.figure().add_subplot(projection='3d')
ax.plot_surface(X,Y,Z,cmap=plt.cm.autumn)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_title('三维曲面')
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1)
y1=np.sin(x)
y2=np.cos(x)

sin_line=plt.plot(x,y1,'r--')
cos_line=plt.plot(x,y2,'b-.')

x1=np.random.randint(0,6,10)
x2=np.random.randint(0,6,10)
y1=np.random.randint(2,5,10)
y2=np.random.randint(2,5,10)

scatter1=plt.scatter(x1,y1,s=20,c='r',marker='*')
scatter2=plt.scatter(x2,y2,s=30,c='b',marker='v')

legend1=plt.legend([sin_line[0],cos_line[0]],
                   ['sin','cos'],
                   loc='lower right')

ax=plt.gca().add_artist(legend1)

legend2=plt.legend([scatter1,scatter2],
                   ['red scatter','blue scatter'],
                   loc='lower right',
                   bbox_to_anchor=(1,0.5)
                  )
plt.show()
```

```python
ax1=plt.subplot(2,4,1)
ax2=plt.subplot(2,4,2,polar=True)
ax3=plt.subplot(2,4,3,polar=True)
ax4=plt.subplot(2,4,4)
ax5=plt.subplot(2,1,2,projection='3d')

plt.tight_layout()
plt.subplots_adjust(wspace=0.5,hspace=0.2)

r=np.arange(1,6,1)
theta=(r-1)*(np.pi/2)
x=np.arange(1,7,0.5)
y=np.linspace(1,3,12)
z=20*np.sin(x+y)

ax1.plot(theta,r,'b--D')
ax2.plot(theta,r,linewidth=3,color='r')
ax3.scatter(theta,r,marker='*',c='g',s=60)
ax4.bar(theta,r)
ax5.plot(x,y,z)

plt.show()
```
