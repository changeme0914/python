```python
####

import math
import numpy as np
import pandas as pd
import matplotlib.pyplot  as plt
m=[0,0.5,1.0,1.5,2.0]
n=[1,2,3,4,5,6,7,8]
X=plt.xticks(m)
Y=plt.yticks(n)
X=np.arange(0,2.1,0.1)
y1=X
y2=X**2
y3=X**3 
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Simple Flot")
plt.plot(X,y1,label="linear")
plt.plot(X,y2,label="quadratic")
plt.plot(X,y3,label="cubic")
plt.legend()
plt.show()
```
